```python
import pandas as pd
import os
```

```python
current_path = os.getcwd()
ov_all_path = os.path.join(current_path,"cellprofier","ov_all","MyExpt_Image.csv")
ov_all = pd.read_csv(ov_all_path)
ov_all.head()
```

```python
ov_all['URL_HEslice'][0]
```

```python
# Extract 'TCGA-04-1331' from 'URL' column and assign to 'Index' column
ov_all['Index'] = ov_all['URL_HEslice'].apply(lambda x: x.split('/')[-3])

# Print first few rows of the DataFrame
print(ov_all["Index"].head())
```

```python
# Get columns that start with 'Median' or 'Mean'
desired_cols = [col for col in ov_all.columns if col.startswith(('Correlation', 
                                                             'Granularity',
                                                             'ImageQuality',
                                                             'Intensity',
                                                             'Texture',
                                                             'Index'))]

# Extract these columns from the DataFrame
ov_all_desired = ov_all[desired_cols]
print(ov_all_desired.shape)
ov_all_desired.head()
```

```python
# Use groupby function and get mean of other columns for each unique Index
ov_all_mean = ov_all_desired.groupby('Index').mean()

print(ov_all_mean.shape)
```

```python
current_path = os.getcwd()
ov_split_path = os.path.join(current_path,"cellprofier","ov_split","MyExpt_Image.csv")
ov_split = pd.read_csv(ov_split_path)
ov_split.head()
```

```python
# Extract 'TCGA-04-1331' from 'URL' column and assign to 'Index' column
ov_split['Index'] = ov_split['URL_HEslice'].apply(lambda x: x.split('/')[-3])

# Get columns that start with 'Median' or 'Mean'
desired_cols = [col for col in ov_split.columns if col.startswith(('Mean', 
                                                             'Median',
                                                             'StDev',
                                                             'Texture',
                                                             'Index'))]

# Extract these columns from the DataFrame
ov_split_desired = ov_split[desired_cols]
print(ov_split_desired.shape)
ov_split_desired.head()
```

```python
# Use groupby function and get mean of other columns for each unique Index
ov_split_mean = ov_split_desired.groupby('Index').mean()

print(ov_split_mean.shape)
```

```python
ov_split_mean.head()
```

```python
# df_merged = pd.merge(df1, df2, on='Index', how='outer')
df_joined = ov_all_mean.join(ov_split_mean, how='outer', lsuffix='_ov_all_mean', rsuffix='_ov_split_mean')
df_joined.shape
```

```python
df_joined.sample(10)
```

```python
df_joined.index
```

```python
current_path = os.getcwd()
ov_all_path = os.path.join(current_path,"files","features_raw.csv")
df_joined.to_csv (ov_all_path) 
```
